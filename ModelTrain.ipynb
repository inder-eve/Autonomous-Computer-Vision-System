```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data="/kaggle/input/aerothon/bike 1/data.yaml",
    epochs=100,          # Keep full 100 epochs
    imgsz=512,           # Slightly smaller image  faster per epoch
    batch=32,            # Maximum batch size for T4 memory
    workers=2,           # Parallel data loading  faster
    project="/kaggle/working",
    name="aerothon_v8n_drone",
    device=0,            # Use GPU
    half=True,           # Mixed precision (FP16)  ~2x speedup
    patience=15          # Early stopping if validation stops improving
)
```
